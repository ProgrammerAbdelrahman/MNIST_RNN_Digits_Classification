```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.utils import to_categorical

# Load the MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Convert the dataset into a binary classification problem
y_train_binary = np.where(y_train == 0, 1, 0)
y_test_binary = np.where(y_test == 0, 1, 0)

# Normalize the pixel values and flatten the images
X_train = X_train / 255.0
X_test = X_test / 255.0
X_train = X_train.reshape((X_train.shape[0], -1))
X_test = X_test.reshape((X_test.shape[0], -1))

# Build a binary classifier
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=784))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train_binary, epochs=10, batch_size=32, validation_data=(X_test, y_test_binary))

# Predict the probabilities of the positive class (digit 0)
y_pred_prob = model.predict(X_test)

# Compute the ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test_binary, y_pred_prob)
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.plot(fpr, tpr, color='blue', label='ROC curve (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='red', linestyle='--', label='Random guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for MNIST Dataset')
plt.legend()
plt.show()
```

```python
# Load the MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize the pixel values and reshape the images
X_train = X_train / 255.0
X_test = X_test / 255.0
X_train = X_train.reshape((X_train.shape[0], -1))
X_test = X_test.reshape((X_test.shape[0], -1))

# Convert the labels to one-hot encoding
y_train_onehot = to_categorical(y_train)
y_test_onehot = to_categorical(y_test)

# Build a multi-class classifier
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=784))
model.add(Dense(10, activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train_onehot, epochs=10, batch_size=32, validation_data=(X_test, y_test_onehot))

# Predict the probabilities for each class
y_pred_prob = model.predict(X_test)

# Compute the ROC curve and AUC for each class
n_classes = y_test_onehot.shape[1]
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_onehot[:, i], y_pred_prob[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Compute the micro-average ROC curve and AUC
fpr_micro, tpr_micro, _ = roc_curve(y_test_onehot.ravel(), y_pred_prob.ravel())
roc_auc_micro = auc(fpr_micro, tpr_micro)

# Plot the ROC curves for each class and the micro-average ROC curve
plt.figure()
plt.plot(fpr_micro, tpr_micro, color='blue', label='Micro-average ROC curve (AUC = %0.2f)' % roc_auc_micro, linestyle='--')
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label='ROC curve of class %d (AUC = %0.2f)' % (i, roc_auc[i]))
plt.plot([0, 1], [0, 1], color='red', linestyle='--', label='Random guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for MNIST Dataset (Multi-class Classification)')
plt.legend()
plt.show()
```
